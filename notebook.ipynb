```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# subset the dataframe for type 'Movie'
netflix_subset= netflix_df[netflix_df['type'] == 'Movie']

# filter the movies of 1990s
movies_1990s = netflix_subset[(netflix_subset['release_year']>=1990)&(netflix_subset['release_year']< 2000)]
```

```python
# using histogram to find most duration of frequency 
plt.hist(movies_1990s['duration'],color = 'red',edgecolor = 'black')
plt.title('Movies duration in 1990s')
plt.xlabel('duaration')
plt.ylabel('Number of movies')
plt.show() 
```

```python
# re-filter the movies of 1990s only with action type
action_movies_1990s = movies_1990s[movies_1990s['genre'] == 'Action']

# count shhort actions movies on action_movies_1990s

short_action_movies_count = 0

for label, row in action_movies_1990s.iterrows():
    if row['duration'] < 90:
        short_action_movies_count = short_action_movies_count + 1
    else:
        short_action_movies_count = short_action_movies_count


print(short_action_movies_count)
```
